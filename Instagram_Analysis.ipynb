```python
import matplotlib.pyplot as plt

top_users = photos["user ID"].value_counts().head(10)

plt.figure(figsize=(8,5))
top_users.plot(kind="bar")
plt.title("Top 10 Most Active Users")
plt.xlabel("User ID")
plt.ylabel("Number of Posts")
plt.savefig("top_10_active_users.png", dpi=300,bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
photos["Insta filter used"].value_counts().head(10).plot(kind="bar", figsize=(8,5))

plt.title("Top 10 Instagram Filters Used")
plt.xlabel("Instagram Filter")
plt.ylabel("Number of posts")
plt.xticks(rotation=45)
plt.savefig("most_used_hashtags.png", dpi=300,bbox_inches="tight")
plt.show()
```

#Business Insight:
The most frequently used filters indicate users preferred editing styles. Brands can use these popular filters to create visually appealing content.

```python
import matplotlib.pyplot as plt

tags["tag text"].value_counts().head(10).plot(kind="bar", figsize=(8,5), color="orange")

plt.title("Top Haashtags")
plt.xlabel("Hashtags")
plt.ylabel("Counts")
plt.xticks(rotation=45)
plt.savefig("posts_by_filter.png", dpi=300,bbox_inches="tight")
plt.show
```

#Business Insight

Popular hashtags help increase post visibility and engagement. Businesses should include trending hashtags in their content strategy

```python
import matplotlib.pyplot as plt

photos["photo type"].value_counts().plot(kind="pie", autopct="%1.1f%%", figsize=(6,6))

plt.title("Distribution of Photo Types")
plt.ylabel("")
plt.savefig("top_liked_photos.png", dpi=300,bbox_inches="tight")
plt.show()
```

#Business Insight:

Understanding the distribution of photo types helps identify which content format users prefer, enabling better content planning
